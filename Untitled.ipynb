```python
#ADF(ada fullar test)
from statsmodels.tsa.stattools import adfuller
```

```python
#Test for staionarity
def test_stationarity(timeseries):
    # Determing rolling statistics
    rolmean = timeseries.rolling(12).mean() # rolling mean
    rolstd = timeseries.rolling(12).std() # rolling standard deviation
    # Plot rolling statistics:
    plt.figure(figsize = (18,8))
    plt.grid('both')
    plt.plot(timeseries, color='blue',label='Original', linewidth = 3)
    plt.plot(rolmean, color='red', label='Rolling Mean',linewidth = 3)
    plt.plot(rolstd, color='black', label = 'Rolling Std',linewidth = 4)
    plt.legend(loc='best', fontsize = 20, shadow=True,facecolor='lightpink',edgecolor = 'k')
    plt.title('Rolling Mean and Standard Deviation', fontsize = 25)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.show(block=False)

    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    # hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
def train_arima_model(X, y, arima_order):
    # prepare training dataset
    # make predictions list
    history = [x for x in X]
    predictions = list()
    for t in range(len(y)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(y[t])
    # calculate out of sample error
    rmse = np.sqrt(mean_squared_error(y, predictions))
    return rmse
```

```python
def evaluate_models(dataset, test, p_values, d_values, q_values):
    dataset=dataset.astype('float32')
    best_score, best_cfg=float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order=(p,d,q)
                try:
                    rmse=train_arima_model(dataset, test, order)
                    if rmse<best_score:
                        best_score, best_cfg=rmse,order
                    print('ARIMA%s RMSE=%.3f' % (order,rmse))
                except:
                    continue
    print('Best ARIMA%s RMSE=%.3f' %(best_cfg,best_score))
```

```python
# evaluate parameters
import warnings
warnings.filterwarnings('ignore')
p_values = range(0, 5)
d_values = range(0, 5)
q_values = range(0, 5)
evaluate_models(train_data,test_data,p_values,d_values,q_values)
```

```python
import itertools
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

# Load your time series data (replace this with your actual data)
# For example, assuming you have a pandas DataFrame 'df' with a 'date' column and a 'value' column:
# df = pd.read_csv('your_data.csv')
# df['date'] = pd.to_datetime(df['date'])
# df.set_index('date', inplace=True)

# Define the range of values for p, d, q, P, D, and Q
p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)
P_values = range(0, 2)
D_values = range(0, 2)
Q_values = range(0, 2)
s = 12  # Seasonal periodicity (you can adjust this based on your data)

# Create all possible combinations of the above parameters
parameter_combinations = list(itertools.product(p_values, d_values, q_values, P_values, D_values, Q_values))

best_model = None
best_mse = float('inf')
history = [x for x in train_data]  # Assuming train_data is your historical time series data

# Iterate through all parameter combinations
for param in parameter_combinations:
    try:
        # Fit SARIMA model
        model = sm.tsa.SARIMAX(history, order=(param[0], param[1], param[2]), 
                                seasonal_order=(param[3], param[4], param[5], s)).fit(disp=False)
        
        # Make predictions
        predictions = model.predict(start=len(history), end=len(history) + 60, dynamic=False)  # Adjust the forecast horizon as needed

        # Update the history variable with actual values from the training data
        history.extend(predictions)

        # Calculate Mean Squared Error (MSE)
        mse = mean_squared_error(history[-len(predictions):], predictions)
        
        # Update the best model if the current model has a lower MSE
        if mse < best_mse:
            best_model = model
            best_mse = mse

    except Exception as e:
        print(f"Error for parameter combination {param}: {e}")

# Print the best model and its parameters
print("Best SARIMA Model:")
print(best_model.summary())
```
